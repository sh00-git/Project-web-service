### 데이터 전처리

#### 결측치 제거

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

#### 데이터 타입 변경

```python
data['신고일'] = pd.to_datetime(data['신고일'])
data['조치일'] = pd.to_datetime(data['조치일'])
```

```python
data.info()
```

#### 특성공학

```python
data['동정보'] = data['주소'].str.split(' ').str[2]
```

```python
data['상세주소'] = data['주소'].str.split(' ').str[3]
```

```python
data['조치소요기간'] = data['조치일'] - data['신고일']
```

```python
data['조치소요기간'] = data['조치소요기간'] / np.timedelta64(1, 'D')
data['조치소요기간']
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data[data['상세주소'].isnull()]
```

```python
data = data.astype({'조치소요기간':'int'})
```

#### db 저장할 데이터프레임

```python
df_db = data[['번호','신고일','조치일','구정보','동정보','상세주소','유형','조치소요기간']]
```

```python
df_db.head()
```

```python
df_db = df_db.where(pd.notnull(df_db), None)
```

## SQLlite 데이터 적재하기

```python
import sqlite3
```

```python
conn = sqlite3.connect("C:/Users/Songhee/bootcamp/Section3/Section3_project/kick_board.db")
cur = conn.cursor()

sql = '''INSERT INTO towing (
id, report_date, action_date, gu, dong, detailed_address, category, action_period)
VALUES(?,?,?,?,?,?,?,?)'''

for index, row in df_db.iterrows():
    report_date = row['신고일']
    print(report_date)
    val = (row['번호'],row['신고일'],row['조치일'],row['구정보'],row['동정보'],row['상세주소'],row['유형'],row['조치소요기간'])
    cur.execute(sql, val)
    print(index)

connection.commit() 

cur.close()
conn.close() 
```
